```python
# import neccessary modules and submodules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing  import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
# show only first four columns
df = pd.read_csv("datasets\\hw.csv")
df
```

```python
# check for whether the dataset have any null values or not
df.isnull().sum()
```

```python
# label encode Gender values
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
df.iloc[:, 0] = enc.fit_transform(df.iloc[:, 0])
df
```

```python
# extract dependent and independent variable
X = df.iloc[:, [1, 2]]
y = df.iloc[:, 0]
```

```python
# visualization of given data
fig = plt.figure(figsize=(28, 20))
ax = plt.axes(projection='3d')
x1_axis = []
x2_axis = []
y1_axis = []
y2_axis = []
z1_axis = []
z2_axis = []
for i in range(len(df)):
    if df.iloc[i, 0] == 0:
        x1_axis.append(df.iloc[i, 0])
        y1_axis.append(df.iloc[i, 1])
        z1_axis.append(df.iloc[i, 2])
    else:
        x2_axis.append(df.iloc[i, 0])
        y2_axis.append(df.iloc[i, 1])
        z2_axis.append(df.iloc[i, 2])
ax.set_xlabel("Gender")
ax.set_ylabel("Height")
ax.set_zlabel("Weight")
ax.scatter3D(x1_axis, y1_axis, z1_axis, c = 'r', s = 1)
ax.scatter3D(x2_axis, y2_axis, z2_axis, c = 'b', s = 1)
ax.set_title('Gender v/s Height v/s Weight', bbox={'facecolor':'0.8', 'pad':5}, loc = 'center')
plt.legend(['Female', 'Male'], ncol=2)
plt.show()
```

```python
# split the dataset and create train and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 42)
```

```python
# logistic regression or binary classifier
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print("Classification Report [Logistic Regression / Binary Classifier]".center(60),"\n\n", classification_report(y_test, y_pred))
```

```python
# k nearest neighbourhood classifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print("Classification Report [K Nearest Neighborhood Classifier]".center(60),"\n\n", classification_report(y_test, y_pred))
```

```python
# decision tree classifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
print("Classification Report [Decision Tree Classifier]".center(60),"\n\n", classification_report(y_test, y_pred))
```

```python
# random forest classifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
print("Classification Report [Random Forest Classifier]".center(60),"\n\n", classification_report(y_test, y_pred))
```

```python
mlpc = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
mlpc.fit(X_train, y_train)
y_pred = mlpc.predict(X_test)
print("Classification Report [Multi Layer Perceptron Classifier]".center(60),"\n\n", classification_report(y_test, y_pred))
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
print("Classification Report [Gaussian Naive Bayes Classifier]".center(60),"\n\n", classification_report(y_test, y_pred))
```

```python
sgdc = SGDClassifier()
sgdc.fit(X_train, y_train)
y_pred = sgdc.predict(X_test)
print("Classification Report [Stochastic Gradient Descent Classifier]".center(60),"\n\n", classification_report(y_test, y_pred))
```
